```python
def preprocess_datasets(dfs):
    for name, df in dfs.items():
        print(f"\nProcessing {name} dataset...")
                
        # Remove duplicates
        df.drop_duplicates(inplace=True)
        
        # Check and handle missing values
        missing_summary = df.isnull().sum()
        print(f"Missing values before handling:\n{missing_summary[missing_summary > 0]}\n")
        
        for col in df.columns:
            if df[col].isnull().sum() > 0:  
                if df[col].dtype == "object":
                    df[col].fillna(df[col].mode()[0], inplace=True)  # Fill categorical with mode
                else:
                    df[col].fillna(df[col].median(), inplace=True)  # Fill numerical with median
    
    print("✅ Preprocessing complete for all datasets!")
    return dfs  # Returns preprocessed DataFrames

# Dictionary of already loaded DataFrames
dfs = {
    "description": description,
    "diets": diets,
    "medications": medicines,
    "precautions": precautions,
    "symptoms": sym_des,
    "training": data,
    "workout": workout
}

# Run preprocessing
preprocessed_data = preprocess_datasets(dfs)

# Example: Check the first few rows of the training dataset
print(preprocessed_data["training"].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Step 1: Prepare your data
X = sym_des[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4']].fillna('')
y = sym_des['Disease']

# Step 2: Combine symptom columns into one string per row (for encoding)
X_combined = X.apply(lambda row: ' '.join(row.values.astype(str)).strip(), axis=1)

# Step 3: Encode symptoms and disease
symptom_encoder = LabelEncoder()
disease_encoder = LabelEncoder()

X_encoded = symptom_encoder.fit_transform(X_combined)
y_encoded = disease_encoder.fit_transform(y)

# Convert to shape (n_samples, 1) to work with scikit-learn
X_encoded = X_encoded.reshape(-1, 1)

# Step 4: Split data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# Step 5: Train and test each model
models = {
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {acc:.2f}")
```
